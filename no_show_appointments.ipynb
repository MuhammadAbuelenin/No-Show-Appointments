## Data Cleaning

```python
# Lets rename so columns and some data as well
df.rename(columns={"Hipertension": "Hypertension"}, inplace=True)
df.rename(columns={"Handcap": "Handicap"}, inplace=True)
df.rename(columns={"No-show": "No_show"}, inplace=True)
df.rename(columns=lambda x: x.strip().lower(), inplace=True) 
```

```python
# Drop all the columns we will not use im suspect about ScheduledDay and AppointmentDay but lets see
df.drop(["patientid", "appointmentid", "scheduledday", "appointmentday"], axis=1, inplace=True)
```

```python
# Now much more simple
df.head()
```

```python
# here we removed the rows with age equal or less than zero around 3540 row
df = df[df["age"] > 0]
print(110527 - df.shape[0])
```

```python
# Assign a new var for no_show which is too confusing
df.no_show.unique()
attend = df.no_show == "No"
didnt_attend = df.no_show == "Yes"
```

```python
# Total No of attending Patients
print("Total Attending:",df[attend].count().gender)
```

```python
# Total No of non attending Patients
print("Total Didnt Attend:",df[didnt_attend].count().gender)
```

```python
# Total patients who recieved and not recieved sms
df.groupby(["sms_received"]).count()["age"]
```

## Lets make our visualization 

```python
# [1] Showing attend and didnt attend according to gender
df.gender[attend].hist(alpha= .75, label="Attend")
df.gender[didnt_attend].hist(alpha= .75, label="Didnt Attend")
plt.legend()
# plt.figure(figsize=(12,12))
plt.xlabel("Gender")
plt.ylabel("Total Patients")
plt.title("Table showing relation between Gender and Attending");
```

```python
print("Attending:-")
print(df.gender[attend].value_counts().to_string())
print("-"*40)
print("Didnt Attend:-")
print(df.gender[didnt_attend].value_counts().to_string())
```

##### Gender is insignificant

```python
# [2] Showing attend and didnt attend according to age
df.age[attend].hist(alpha= .75, label="Attend")
df.age[didnt_attend].hist(alpha= .75, label="Didnt Attend")
plt.legend()
plt.xlabel("Age")
plt.ylabel("Total Patients")
plt.title("Table showing relation between Age and Attending");
```

##### Age is insignificant

```python
# [3] Showing attend and didnt attend according to neighbourhood
plt.figure(figsize=[18,12])
df.neighbourhood[attend].value_counts().plot(kind="bar", alpha= .75, label="Attend", color="blue")
df.neighbourhood[didnt_attend].value_counts().plot(kind="bar",alpha= .75,label="Didnt Attend",color="orange")
plt.legend()
plt.xlabel("Neighbourhood")
plt.ylabel("Total Patients")
plt.title("Table showing relation between Neighbourhood and Attending");
```

```python
df.groupby(["neighbourhood", "no_show"], group_keys=False).sum()["age"]
```

```python
# Most attending patient neighbourhood
most_neighbourhood = df.neighbourhood.mode()[0]
print(most_neighbourhood,"=", df.query('neighbourhood == "JARDIM CAMBURI"')[attend].count()["no_show"])
```

#### Here we can find a strong relation between attending and one Neighbourhood which high probability hospital neighbourhood
